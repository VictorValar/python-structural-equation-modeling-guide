```python
import pyreadstat

df, meta = pyreadstat.read_sav("../data/EXERCICIO_2025.sav")
df.head()
```

```python
cols = [
    'P1.1_QA', 'P1.2_QA', 'P1.3_QA', 'P1.4_QA', 'P1.5_QA',
    'P1.6_QC', 'P1.7_QC', 'P1.8_QC', 'P1.9_QC',
    'P1.10_QT', 'P1.11_QT', 'P1.12_QT', 'P1.13_QT', 'P1.14_QT',
    'P1.15_QA', 'P1.16_QA', 'P1.17_QA',
    'P3.1_S', 'P3.2_S', 'P3.3_S', 'P3.4_S',
    'P6.1_L', 'P6.2_L', 'P6.3_L', 'P6.4_L', 'P6.5_L', 'P6.6_L'
]

clean_df = df[cols].copy()

clean_df.rename(columns=lambda c: c.replace('.', '_'), inplace=True)
clean_df.head()
```

```python
from semopy import Model

# Atendimento = QA
# Financeiro = QF
# QualidadeTecnica = QT
# Satisfacao = S
# Lealdade = L
# QA =~ P1.1_QA + P1.2_QA + P1.3_QA + P1.4_QA + P1.5_QA + P1.15_QA + P1.16_QA + P1.17_QA
# QF =~ P1.6_QC + P1.7_QC + P1.8_QC + P1.9_QC
# QT =~ P1.10_QT + P1.11_QT + P1.12_QT + P1.13_QT + P1.14_QT
# S =~ P3.1_S + P3.2_S + P3.3_S + P3.4_S
# L =~ P6.1_L + P6.2_L + P6.3_L + P6.4_L + P6.5_L + P6.6_L

mod = """
QA =~ P1_1_QA + P1_3_QA + P1_4_QA + P1_5_QA + P1_15_QA
QF =~ P1_6_QC + P1_7_QC + P1_8_QC
QT =~ P1_10_QT + P1_11_QT + P1_12_QT + P1_13_QT
"""

model = Model(mod)
model.fit(clean_df)
```

χ²/df < 5

CFI, GFI, TLI > 0.9

RMSEA < 0,08

```python
from semopy import calc_stats

stats = calc_stats(model)
stats
```

```python
print("Default Model:", stats['chi2'] / stats['DoF'])
print("Independence Model:", stats['chi2 Baseline'] / stats['DoF Baseline'])
```

```python
# Regression Weights
estimates = (
    model.inspect(std_est=True)
         .query("op == '~' and lval.str.startswith('P')", engine='python')
)
estimates
```

```python
latent_scores = model.predict_factors(clean_df)
cor_matrix = latent_scores.corr()

print(cor_matrix)
```

### Variância Média Extraída (AVE)
Simplified formula

$$
\text{AVE} \;=\;
\frac{\displaystyle\sum \lambda^{2}}
     {\displaystyle\sum \lambda^{2} \;+\; \sum \epsilon}
$$


### Confiabilidade Composta (CR)
Simplified formula

$$
\text{CR} \;=\;
\frac{\displaystyle\left(\sum \lambda\right)^{2}}
     {\displaystyle\left(\sum \lambda\right)^{2} \;+\; \sum \epsilon}
$$

R2 > 0.2


```python
standardized_loadings = (
    model.inspect(std_est=True)
    .query("op == '~' and lval.str.startswith('P')", engine='python')
    .loc[:, ['lval', 'rval', 'Est. Std']]  # Item, Fator, λ
    .rename(columns={'lval': 'Item', 'rval': 'Fator', 'Est. Std': 'lambda'})
)
standardized_loadings["lambdaˆ2"] = standardized_loadings["lambda"].apply(lambda x: x ** 2)
standardized_loadings["epsilon"] = standardized_loadings["lambdaˆ2"].apply(lambda x: 1 - x)
standardized_loadings
```
